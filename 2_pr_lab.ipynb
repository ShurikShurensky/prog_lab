```python
# Leetcode: 20. Valid Parentheses

def isValid(s: str) -> bool:
    stack = []
    for ch in s:
        if ch in "({[":
            stack.append(ch)
        else:
            if not stack:
                return False
            top = stack.pop()
            if ch == ")" and top != "(":
                return False
            if ch == "}" and top != "{":
                return False
            if ch == "]" and top != "[":
                return False
    return not stack
```

```python
# Leetcode: 150. Evaluate Reverse Polish Notation

from typing import List

def evalRPN(tokens: List[str]) -> int:
    stack = []
    for token in tokens:
        if token in "+-*/":
            a = stack.pop()         
            b = stack.pop()
            if token == "+":
                stack.append(b + a)
            elif token == "-":
                stack.append(b - a)
            elif token == "*":
                stack.append(b * a)
            else:  # "/"
                stack.append(int(b / a))
        else:
            stack.append(int(token))
    return stack[0]

tokens = ["10","6","9","3","+","-11","*","/","*","17","+","5","+"]
print(evalRPN(tokens))  
```

```python
# Leetcode: 22. Generate Parentheses

def generateParenthesis(n: int) -> list[str]:
    result = []
    
    stack = [("", 0, 0)]
    
    while stack:
        current, open_count, close_count = stack.pop()
        
        if open_count == n and close_count == n:
            result.append(current)
            continue
        if close_count < open_count:
            stack.append((current + ")", open_count, close_count + 1))
        
        if open_count < n:
            stack.append((current + "(", open_count + 1, close_count))
    
    return result

print(generateParenthesis(3))
```

```python
# Leetcode 232. Implement Queue using Stacks

class MyQueue:

    def __init__(self):
        self.in_stack = []
        self.out_stack = []

    def push(self, x: int) -> None:
        self.in_stack.append(x)

    def pop(self) -> int:
        if not self.out_stack:
            while self.in_stack:
                self.out_stack.append(self.in_stack.pop())
        return self.out_stack.pop()

    def peek(self) -> int:
        if not self.out_stack:
            while self.in_stack:
                self.out_stack.append(self.in_stack.pop())
        return self.out_stack[-1]

    def empty(self) -> bool:
        return not self.in_stack and not self.out_stack

```

```python
# Leetcode 622. Design Circular Queue

class MyCircularQueue:
    def __init__(self, k: int):
        self.capacity = k
        self.queue = [0] * k  # fixed-size array to hold elements
        self.head = 0
        self.tail = 0
        self.count = 0

    def enQueue(self, value: int) -> bool:
        if self.isFull():
            return False
        self.queue[self.tail] = value
        self.tail = (self.tail + 1) % self.capacity
        self.count += 1
        return True

    def deQueue(self) -> bool:
        if self.isEmpty():
            return False
        self.head = (self.head + 1) % self.capacity
        self.count -= 1
        return True

    def Front(self) -> int:
        if self.isEmpty():
            return -1
        return self.queue[self.head]

    def Rear(self) -> int:
        if self.isEmpty():
            return -1
        # tail points to the next insertion index, so the last element is at tail - 1
        return self.queue[(self.tail - 1 + self.capacity) % self.capacity]

    def isEmpty(self) -> bool:
        return self.count == 0

    def isFull(self) -> bool:
        return self.count == self.capacity
```
